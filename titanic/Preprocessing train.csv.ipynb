```python
# train.csv 파일을 전처리 하기위한 스크립트코드.
import pandas as pd #pandas DataFrame 처리를 위한 패키지
import numpy as np #python에서 수학식을 처리하기 위한 패키지

#train.csv 로드
data = pd.read_csv('/Zhim/2020/train.csv')
dummy_data = data #원본 훼손을 방지하기 위해 변경될 데이터 프레임을 생성

#학습에 필요없다고 생각되는 열 삭제
del dummy_data['Name']
del dummy_data['Cabin']
del dummy_data['Fare']
del dummy_data['Ticket']

#생존확률을 표시할 Survival probability(%) 열을 추가
dummy_data['Survival probability(%)'] = 0

#Age(나이) 열에서 NAN 값을 나이 평균값으로 대체
dummy_data['Age'].fillna(value = dummy_data['Age'].mean())
#dummy_data.loc[dummy_data["Age"]=="Nan","Age"] = dummy_data['Age'].mean()

#Sex(성별을) 남자 : 0 , 여자 : 1로 인코딩
dummy_data.loc[dummy_data["Sex"] == "male","Sex"] = 0
dummy_data.loc[dummy_data["Sex"] == "female","Sex"] = 1
```
